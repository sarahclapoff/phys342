# Gaussian distribution

```python
import numpy as np
from scipy import stats
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
mean = 1; standardDeviation = 2; dxMax = 6.
xArray = np.arange(mean-dxMax,mean+dxMax,dxMax/100)
probabilityDensity = stats.norm.pdf(xArray,loc=mean,scale=standardDeviation)
plt.plot(xArray,probabilityDensity);
plt.title("Gaussian Distribution"); plt.xlabel("x"); plt.ylabel("probability density")
plt.show()
```
